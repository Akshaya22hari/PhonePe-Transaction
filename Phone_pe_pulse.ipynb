Agg Transaction


```python

#Required libraries for the program

import pandas as pd
import json
import os

#This is to direct the path to get the data as states

path="/content/pulse/data/aggregated/transaction/country/india/state/"
Agg_state_list=os.listdir(path)
Agg_state_list
#Agg_state_list--> to get the list of states in India

#This is to extract the data's to create a dataframe

clm={'State':[], 'Year':[],'Quater':[],'Transacion_type':[], 'Transacion_count':[], 'Transacion_amount':[]}

for i in Agg_state_list:
    p_i=path+i+"/"
    Agg_yr=os.listdir(p_i)
    for j in Agg_yr:
        p_j=p_i+j+"/"
        Agg_yr_list=os.listdir(p_j)
        for k in Agg_yr_list:
            p_k=p_j+k
            Data=open(p_k,'r')
            D=json.load(Data)
            for z in D['data']['transactionData']:
              Name=z['name']
              count=z['paymentInstruments'][0]['count']
              amount=z['paymentInstruments'][0]['amount']
              clm['Transacion_type'].append(Name)
              clm['Transacion_count'].append(count)
              clm['Transacion_amount'].append(amount)
              clm['State'].append(i)
              clm['Year'].append(j)
              clm['Quater'].append(int(k.strip('.json')))
#Succesfully created a dataframe
Agg_Transaction=pd.DataFrame(clm)
```

Aggregrated User

```python

#This is to direct the path to get the data as states

user_path="/content/pulse/data/aggregated/user/country/india/state/"
# Define insurance_path - it should point to the directory containing state-level data

Agg_state_list = os.listdir(user_path)
Agg_state_list


clm_user = {'State': [], 'Year': [], 'Quater': [], 'Brand': [], 'User_count': [], 'User_percentage': []}

for i in Agg_state_list:
    p_i = user_path + i + "/"  # Corrected the path construction
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        p_j = p_i + j + "/"
        Agg_yr_list = os.listdir(p_j)
        for k in Agg_yr_list:
            p_k = p_j + k
            Data = open(p_k, 'r')
            D = json.load(Data)
            # Check if 'data' and 'usersByDevice' keys exist before iterating
            if D['data'] is not None and 'usersByDevice' in D['data'] and D['data']['usersByDevice']:
                for z in D['data']['usersByDevice']:
                    brand = z['brand']
                    count = z['count']
                    percentage = z['percentage']
                    clm_user['Brand'].append(brand)
                    clm_user['User_count'].append(count)
                    clm_user['User_percentage'].append(percentage)
                    clm_user['State'].append(i)
                    clm_user['Year'].append(j)
                    clm_user['Quater'].append(int(k.strip('.json')))

# Successfully created a dataframe
Agg_User = pd.DataFrame(clm_user)
```

Aggregrated Insurance

```python


#This is to direct the path to get the data as states

#path="/content/pulse/data/aggregated/insurance/country/india/"
# Define insurance_path here - it should point to the directory containing state-level data
insurance_path = "/content/pulse/data/aggregated/insurance/country/india/"


Agg_state_list = os.listdir(insurance_path) # Changed to insurance_path
Agg_state_list
#Agg_state_list--> to get the list of states in India

# This is to extract the data's to create a dataframe for Aggregated Insurance

clm_insurance = {'State': [], 'Year': [], 'Quater': [], 'Insurance_type': [], 'Insurance_count': [], 'Insurance_amount': []}

for i in Agg_state_list:
    # Use insurance_path to construct the file path
    p_i = insurance_path + i + "/"
    Agg_yr = os.listdir(p_i)
    for j in Agg_yr:
        # The following condition is added to check if it is a file instead of a directory
        # If it's a file, we directly process it. Otherwise, we skip it.
        if os.path.isfile(p_i + j):
            p_k = p_i + j
            Data = open(p_k, 'r')
            D = json.load(Data)
            if D['data'] is not None and 'transactionData' in D['data']:
                for z in D['data']['transactionData']:
                    Name = z['name']
                    count = z['paymentInstruments'][0]['count']
                    amount = z['paymentInstruments'][0]['amount']
                    clm_insurance['Insurance_type'].append(Name)
                    clm_insurance['Insurance_count'].append(count)
                    clm_insurance['Insurance_amount'].append(amount)
                    clm_insurance['State'].append(i)
                    clm_insurance['Year'].append(j.strip('.json')) # Extract year from the filename
                    # Assuming the quarter is not available in this file structure, setting it to None
                    clm_insurance['Quater'].append(None)
        else:
            continue  # Skip if it is not a file

# Successfully created a dataframe
Agg_Insurance = pd.DataFrame(clm_insurance)
```

```python
map_user_path = "/content/pulse/data/map/user/hover/country/india/state/"
Map_state_list = os.listdir(map_user_path)
Map_state_list

# This is to extract the data's to create a dataframe for Map User

# Corrected Map User extraction based on 'hoverData'

clm_map_user = {'State': [], 'Year': [], 'Quater': [], 'District': [], 'RegisteredUsers': [], 'AppOpens': []}

for i in Map_state_list:
    p_i = map_user_path + i + "/"
    Map_yr = os.listdir(p_i)
    for j in Map_yr:
        p_j = p_i + j + "/"
        Map_yr_list = os.listdir(p_j)
        for k in Map_yr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D = json.load(Data)
                if D['data'] is not None and 'hoverData' in D['data']:
                    for district_name, district_data in D['data']['hoverData'].items():
                        registeredUsers = district_data.get('registeredUsers', 0)
                        appOpens = district_data.get('appOpens', 0)
                        clm_map_user['District'].append(district_name)
                        clm_map_user['RegisteredUsers'].append(registeredUsers)
                        clm_map_user['AppOpens'].append(appOpens)
                        clm_map_user['State'].append(i)
                        clm_map_user['Year'].append(j)
                        clm_map_user['Quater'].append(int(k.strip('.json')))

# Successfully created a dataframe
Map_User = pd.DataFrame(clm_map_user)

```

```python
top_user_path = "/content/pulse/data/top/user/country/india/state/"
top_state_list = os.listdir(top_user_path)
top_state_list

# This is to extract the data's to create a dataframe for Top User

clm_top_user = {'State': [], 'Year': [], 'Quater': [], 'District_Pincode': [], 'RegisteredUsers': [], 'Type': []}

for i in top_state_list:
    p_i = top_user_path + i + "/"
    Top_yr = os.listdir(p_i)
    for j in Top_yr:
        p_j = p_i + j + "/"
        Top_yr_list = os.listdir(p_j)
        for k in Top_yr_list:
            p_k = p_j + k
            with open(p_k, 'r') as Data:
                D = json.load(Data)
                if D['data'] is not None:
                    # Extract Districts
                    if 'districts' in D['data'] and D['data']['districts'] is not None:
                        for district in D['data']['districts']:
                            district_name = district['name']
                            registeredUsers = district['registeredUsers']
                            clm_top_user['District_Pincode'].append(district_name)
                            clm_top_user['RegisteredUsers'].append(registeredUsers)
                            clm_top_user['Type'].append('District')
                            clm_top_user['State'].append(i)
                            clm_top_user['Year'].append(j)
                            clm_top_user['Quater'].append(int(k.strip('.json')))
                    # Extract Pincodes
                    if 'pincodes' in D['data'] and D['data']['pincodes'] is not None:
                        for pincode in D['data']['pincodes']:
                            pincode_name = pincode['name']
                            registeredUsers = pincode['registeredUsers']
                            clm_top_user['District_Pincode'].append(pincode_name)
                            clm_top_user['RegisteredUsers'].append(registeredUsers)
                            clm_top_user['Type'].append('Pincode')
                            clm_top_user['State'].append(i)
                            clm_top_user['Year'].append(j)
                            clm_top_user['Quater'].append(int(k.strip('.json')))

# Successfully created a dataframe
Top_User = pd.DataFrame(clm_top_user)
```

```python
# Create Aggregated Tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS Aggregated_user (
    State TEXT,
    Year INTEGER,
    Quater INTEGER,
    User_type TEXT,
    User_count INTEGER
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Aggregated_transaction (
    State TEXT,
    Year INTEGER,
    Quater INTEGER,
    Transaction_type TEXT,
    Transaction_count INTEGER,
    Transaction_amount REAL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Aggregated_insurance (
    State TEXT,
    Year INTEGER,
    Quater INTEGER,
    Insurance_type TEXT,
    Insurance_count INTEGER,
    Insurance_amount REAL
);
''')

# Create Map Tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS Map_user (
    State TEXT,
    District_Pincode TEXT,
    User_count INTEGER,
    User_amount REAL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Map_transaction (
    State TEXT,
    District_Pincode TEXT,
    Transaction_type TEXT,
    Transaction_count INTEGER,
    Transaction_amount REAL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Map_insurance (
    State TEXT,
    District_Pincode TEXT,
    Insurance_count INTEGER,
    Insurance_amount REAL
);
''')

# Create Top Tables
cursor.execute('''
CREATE TABLE IF NOT EXISTS Top_user (
    State TEXT,
    Year INTEGER,
    Quater INTEGER,
    User_type TEXT,
    User_count INTEGER
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Top_transaction (
    State TEXT,
    Year INTEGER,
    Quater INTEGER,
    Transaction_type TEXT,
    Transaction_count INTEGER,
    Transaction_amount REAL
);
''')

cursor.execute('''
CREATE TABLE IF NOT EXISTS Top_insurance (
    State TEXT,
    Year INTEGER,
    Quater INTEGER,
    Insurance_type TEXT,
    Insurance_count INTEGER,
    Insurance_amount REAL
);
''')

# Commit the changes and close the connection
#conn.commit()
```

```python
#Total Transaction Amount & Count — State-wise
# Group by State
state_summary = df.groupby('State').agg({
    'Transaction_count': 'sum',
    'Transaction_amount': 'sum'
}).sort_values('Transaction_amount', ascending=False)

state_summary
```

```python
#Category-wise Analysis (Transaction_type)
# Group by Transaction_type
category_summary = df.groupby('Transaction_type').agg({
    'Transaction_count': 'sum',
    'Transaction_amount': 'sum'
}).sort_values('Transaction_amount', ascending=False)

category_summary
```

```python
#Identify Declining States or Categories
# Compare growth between Years
yearly_summary = df.groupby(['State', 'Year']).agg({'Transaction_amount':'sum'}).reset_index()

# Pivot to easier format
pivot_table = yearly_summary.pivot(index='State', columns='Year', values='Transaction_amount')

pivot_table['Growth'] = (pivot_table[2023] - pivot_table[2022]) / pivot_table[2022] * 100
pivot_table.sort_values('Growth')
```
